```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_wine
```

```python
# Load the wine dataset
wine = load_wine()
X = wine.data[:, [9, 12]]  # Select features 9 and 12 for simplicity
```

```python
# Perform K-means clustering
k = 3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)

# Get the cluster labels for K-means
kmeans_labels = kmeans.labels_
```

```python
# Perform Gaussian Mixture Model clustering using EM algorithm
gmm = GaussianMixture(n_components=k)
gmm.fit(X)

# Get the cluster labels for GMM
gmm_labels = gmm.predict(X)
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot K-means clustering
axes[0].scatter(X[kmeans_labels == 0, 0], X[kmeans_labels == 0, 1], label='Cluster 1')
axes[0].scatter(X[kmeans_labels == 1, 0], X[kmeans_labels == 1, 1], label='Cluster 2')
axes[0].scatter(X[kmeans_labels == 2, 0], X[kmeans_labels == 2, 1], label='Cluster 3')
axes[0].set_title('K-means Clustering')
axes[0].set_xlabel('Flavanoids')
axes[0].set_ylabel('Proline')
axes[0].legend()

# Plot EM Algorithm - Gaussian Mixture Model
axes[1].scatter(X[gmm_labels == 0, 0], X[gmm_labels == 0, 1], label='Cluster 1')
axes[1].scatter(X[gmm_labels == 1, 0], X[gmm_labels == 1, 1], label='Cluster 2')
axes[1].scatter(X[gmm_labels == 2, 0], X[gmm_labels == 2, 1], label='Cluster 3')
axes[1].set_title('EM Algorithm - Gaussian Mixture Model')
axes[1].set_xlabel('Flavanoids')
axes[1].set_ylabel('Proline')
axes[1].legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
