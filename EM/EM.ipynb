```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_wine
```

```python
# Load the wine dataset
wine = load_wine()
X = wine.data[:, [9, 12]]  # Select features 9 and 12 for simplicity
```

```python
# Perform Gaussian Mixture Model clustering using EM algorithm
k = 3
gmm = GaussianMixture(n_components=k)
gmm.fit(X)
```

```python
# Get the cluster labels
labels = gmm.predict(X)
```

```python
# Plot the data points and their cluster assignments
plt.figure(figsize=(8, 6))
for i in range(k):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i+1}')
plt.title('EM Algorithm - Gaussian Mixture Model')
plt.xlabel('Flavanoids')
plt.ylabel('Proline')
plt.legend()
plt.show()
```
